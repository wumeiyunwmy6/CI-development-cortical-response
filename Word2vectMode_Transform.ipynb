```python
from transformers import BertTokenizer, BertModel
import torch

# 加载预训练 RoBERTa 模型和分词器
model_name = "hfl/chinese-bert-wwm-ext" # hfl/chinese-roberta-wwm-ext/hfl/chinese-roberta-wwm-ext-large/hfl/chinese-bert-wwm-ext/hfl/chinese-bert-wwm
##okenizer = AutoTokenizer.from_pretrained(model_name)
#model = AutoModel.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
# 准备输入文本
text = "从前有位皇帝，非常喜欢穿好看的新衣服，他把所有的钱都花在做新衣服买衣服上。他一点儿也不关心老百姓，他几乎每一个钟点都要换一套衣服，所以只要有人来找他，回答总是皇上在更衣室换衣服。"

# 使用分词器将文本拆分为单个字
tokens = tokenizer(list(text), return_tensors="pt", is_split_into_words=True)

# 获取模型的输出
outputs = model(**tokens)

# 获取每个字的词向量
word_embeddings = outputs.last_hidden_state
```
